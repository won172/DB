```python
import pandas as pd

# ğŸ“Œ ì‚¬ìš©í•  íŒŒì¼ ëª©ë¡ (2016ë…„ 1ì›”ë¶€í„° ë°ì´í„°ê°€ ìˆëŠ” ETFë§Œ)
file_paths = {
    "S&P500": "^GSPC.csv",
    "ERTH": "ERTH.csv",
    "VDE": "VDE.csv",
    "TAN": "TAN.csv",
    "SUSA": "SUSA.csv"
}

# ğŸ“Œ ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
dataframes = {}
for etf, path in file_paths.items():
    df = pd.read_csv(path, index_col=0, parse_dates=True)
    df = df[['Close']]  # ì¢…ê°€ë§Œ ì‚¬ìš©
    df.rename(columns={'Close': etf}, inplace=True)
    dataframes[etf] = df

# ğŸ“Œ ëª¨ë“  ë°ì´í„° í•©ì¹˜ê¸°
merged_data = pd.concat(dataframes.values(), axis=1, join="inner")

# âœ… ë°ì´í„° í™•ì¸
merged_data.head()
```

```python
# ğŸ“Œ ë¬¸ìì—´ ë°ì´í„°ë¥¼ floatë¡œ ë³€í™˜
merged_data = merged_data.apply(pd.to_numeric, errors='coerce')

# ğŸ“Œ ë³€í™˜ í›„ ë°ì´í„° í™•ì¸
print("\nğŸ“Œ ë³€í™˜ëœ ë°ì´í„° ìƒ˜í”Œ:")
merged_data.head()
```

```python
# ğŸ“Œ ì²« ë‘ í–‰ ì œê±° (ë¶ˆí•„ìš”í•œ í–‰ ì œê±°)
merged_data = merged_data.iloc[2:].reset_index(drop=True)

# ğŸ“Œ ì²« ë²ˆì§¸ ì»¬ëŸ¼ëª…ì„ 'Date'ë¡œ ì„¤ì •
merged_data.rename(columns={merged_data.columns[0]: "Date"}, inplace=True)

# ğŸ“Œ 'Date' ì»¬ëŸ¼ì„ ë‚ ì§œ í˜•ì‹ìœ¼ë¡œ ë³€í™˜
merged_data["Date"] = pd.to_datetime(merged_data["Date"])

# ğŸ“Œ 'Date' ì»¬ëŸ¼ì„ ì¸ë±ìŠ¤ë¡œ ì„¤ì •
merged_data.set_index("Date", inplace=True)

# âœ… ì •ë¦¬ëœ ë°ì´í„° í™•ì¸
print("\nğŸ“Œ ì •ë¦¬ëœ ë°ì´í„° ìƒ˜í”Œ:")
merged_data.head()
```

```python
# ğŸ“Œ ì¼ë°˜ìˆ˜ìµë¥ (Arithmetic Return) ê³„ì‚°
simple_returns = merged_data.pct_change().dropna()

# âœ… ìˆ˜ìµë¥  ë°ì´í„° í™•ì¸
print("\nğŸ“Œ ì¼ë°˜ìˆ˜ìµë¥  ìƒ˜í”Œ:")
simple_returns.head()
```

```python
import statsmodels.api as sm

# ğŸ“Œ ì¶”ì • ìœˆë„ìš° ì„¤ì •
estimation_start = "2016-01-06"
estimation_end = "2016-12-30"

# ğŸ“Œ ë²¤ì¹˜ë§ˆí¬(S&P500) ìˆ˜ìµë¥ 
market_returns = simple_returns["S&P500"]

# ğŸ“Œ ETFë³„ Î±(ì•ŒíŒŒ), Î²(ë² íƒ€) ì¶”ì •
alpha_beta = {}

for etf in simple_returns.columns:
    if etf == "S&P500":
        continue  # ë²¤ì¹˜ë§ˆí¬ëŠ” ì œì™¸

    etf_returns = simple_returns[etf]

    # ğŸ“Œ ì¶”ì • ê¸°ê°„ì˜ ë°ì´í„° ì„ íƒ
    estimation_index = (market_returns.index >= estimation_start) & (market_returns.index <= estimation_end)
    X = market_returns.loc[estimation_index]
    Y = etf_returns.loc[estimation_index]

    # ğŸ“Œ OLS íšŒê·€ ë¶„ì„
    X = sm.add_constant(X)  # ìƒìˆ˜í•­ ì¶”ê°€ (Î± ì¶”ì •)
    model = sm.OLS(Y, X).fit()

    # ê²°ê³¼ ì €ì¥
    alpha, beta = model.params
    alpha_beta[etf] = {"alpha": alpha, "beta": beta}

# âœ… Î±, Î² ê°’ í™•ì¸
print("\nğŸ“Œ ETFë³„ Î±(ì•ŒíŒŒ), Î²(ë² íƒ€) ê°’:")
for etf, params in alpha_beta.items():
    print(f"{etf}: Î± = {params['alpha']:.6f}, Î² = {params['beta']:.6f}")
```
