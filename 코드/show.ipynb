```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ETF 
etf_data = {
    "ESGU": [110.17793, 102.461235, 99.99598, 99.860214],
    "ERTH": [42.148796, 42.01829, 42.003017, 42.00285],
    "NULG": [47.1812, 43.726456, 43.228065, 43.22003],
    "VDE": [89.87605, 83.565956, 81.12938, 80.93898],
    "TAN": [71.53256, 83.10222, 85.1236, 85.16748],
    "SUSA": [95.11496, 89.99996, 88.98026, 88.95337],
}

days = np.array([50, 100, 200, 300])

#  
plt.figure(figsize=(13, 10))

#  ETF  
for etf, prices in etf_data.items():
    plt.plot(days, prices, marker='o', linestyle='-', linewidth=2, markersize=6, label=etf)

#  
plt.xlabel("Days Ahead", fontsize=12)
plt.ylabel("Predicted Price", fontsize=12)
plt.title("Predicted Prices of Various ETFs Over Time", fontsize=14)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.axvline(50, color='red', linestyle='--', linewidth=1)
plt.axvline(100, color='red', linestyle='--', linewidth=1)
plt.axvline(200, color='red', linestyle='--', linewidth=1)
plt.axvline(300, color='red', linestyle='--', linewidth=1)


#  
plt.show()
```

```python
#   (: 50  )
returns_data = {etf: [(price / prices[0] - 1) * 100 for price in prices] for etf, prices in etf_data.items()}

#  
plt.figure(figsize=(20, 10))

#  ETF  
for etf, returns in returns_data.items():
    plt.plot(days, returns, marker='o', linestyle='-', linewidth=2, markersize=6, label=etf)

#  
plt.xlabel("Days Ahead", fontsize=12)
plt.ylabel("Predicted Return (%)", fontsize=12)
plt.title("Predicted Returns of Various ETFs Over Time", fontsize=14)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend()
plt.axhline(y=0, color='black', linestyle='--', linewidth=1)  # 0%   
plt.grid(True, linestyle="--", alpha=0.6)
plt.axvline(50, color='red', linestyle='--', linewidth=1)
plt.axvline(100, color='red', linestyle='--', linewidth=1)
plt.axvline(200, color='red', linestyle='--', linewidth=1)
plt.axvline(300, color='red', linestyle='--', linewidth=1)

#  
plt.show()
```
