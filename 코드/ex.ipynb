```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
import platform

# 한글 폰트 설정
if platform.system() == 'Windows':
    plt.rc('font', family='Malgun Gothic')  # Windows의 경우
elif platform.system() == 'Darwin':  # macOS
    plt.rc('font', family='AppleGothic')
else:
    plt.rc('font', family='NanumGothic')  # Linux 시스템에서는 나눔고딕 사용

# 유니코드 마이너스 기호 사용 설정
plt.rc('axes', unicode_minus=False)
import warnings
warnings.filterwarnings('ignore')

# 데이터 불러오기
file_path = 'C:/Users/swc08/Desktop/대학/DB보험금융공모전/미래에셋지속가능ESG채권증권자투자신탁(채권) 수익률.csv'
data = pd.read_csv(file_path)

# 컬럼 이름 변경
data.columns = ['날짜', '수익률']

# 날짜를 datetime 형식으로 변환
data['날짜'] = pd.to_datetime(data['날짜'])

# 시계열 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(data['날짜'], data['수익률'],  linestyle='-', color='red', markersize=3)
plt.title('미래에셋 지속가능 ESG 채권 수익률')
plt.xlabel('날짜')
plt.ylabel('수익률 (%)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import yfinance as yf
import numpy as np
import pandas as pd

# ESGU 데이터 다운로드
ticker = "ESGU"
data = yf.download(ticker, start="2020-01-01", end="2023-12-31")

# 로그 수익률 계산
data['Log Returns'] = np.log(data['Close'] / data['Close'].shift(1))

# 일별 변동성 계산
daily_volatility = data['Log Returns'].std()

# 연율화된 변동성 계산
annualized_volatility = daily_volatility * np.sqrt(252)

print(f"Daily Volatility: {daily_volatility:.4%}")
print(f"Annualized Volatility: {annualized_volatility:.4%}")
```

```python
import matplotlib.pyplot as plt

# 이동 표준편차로 변동성 추세 계산
data['30-Day Rolling Volatility'] = data['Log Returns'].rolling(window=30).std() * np.sqrt(252)

# 변동성 시각화
plt.figure(figsize=(12, 6))
plt.plot(data['30-Day Rolling Volatility'], label="30-Day Rolling Volatility")
plt.title("ESGU Volatility (30-Day Rolling)")
plt.xlabel("Date")
plt.ylabel("Annualized Volatility")
plt.legend()
plt.grid()
plt.show()
```
